```python
import pandas as pd
import matplotlib.pyplot as plt
flights_data['SnowDepth_Category'] = flights_data['DailySnowDepth_ORIGIN'].apply(lambda x: 'Snow > 0' if x > 0 else 'Snow = 0')
flights_data['Cancelled'] = pd.to_numeric(flights_data['Cancelled'], errors='coerce')

cancelled_snow_proportion = flights_data.groupby(['Cancelled', 'SnowDepth_Category']).size().unstack(fill_value=0)
cancelled_snow_proportion = cancelled_snow_proportion.div(cancelled_snow_proportion.sum(axis=1), axis=0)  

plt.figure(figsize=(12, 6))
cancelled_snow_proportion.plot(kind='bar', stacked=True, color=['skyblue', 'salmon'])


plt.title('Proportion of origin snow depth > 0 by Cancellation Status', fontsize=16)
plt.xlabel('Cancellation Status', fontsize=12)
plt.ylabel('Proportion', fontsize=12)

plt.legend(title='Snow Depth', labels=['Snow = 0', 'Snow > 0'])
plt.tight_layout()
plt.show()
```
