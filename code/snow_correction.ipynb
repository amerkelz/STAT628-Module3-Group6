# Correcting snow data

## 1. Load final dataset,  add correct snow data to a dataframe

```python
import pandas as pd
flights_data=pd.read_csv("flight_weather.csv")
```

```python
flights_data = flights_data.drop(columns=['DailySnowDepth_ORIGIN', 'DailySnowfall_ORIGIN', 'DailySnowDepth_DEST', 'DailySnowfall_DEST'])
```

```python
import os
import pandas as pd
import glob
output_dir = "downloaded_csv_files"
final_df = pd.DataFrame()
for folder in os.listdir(output_dir):
    folder_path = os.path.join(output_dir, folder)
    if os.path.isdir(folder_path):
        folder_df = pd.DataFrame()
        
        for csv_file in glob.glob(os.path.join(folder_path, "*.csv")):
            try:
                df = pd.read_csv(csv_file)
                filtered_df = df[df["REPORT_TYPE"] == "SOD"]
                folder_df = pd.concat([folder_df, filtered_df], ignore_index=True)
            except Exception as e:
                print(f"Error reading {csv_file}: {e}")
        if not folder_df.empty:
            folder_df["Location"] = folder  
            final_df = pd.concat([final_df, folder_df], ignore_index=True)

print(final_df)
```

## 2.merge the flights_data with snow dataframe

```python
import pandas as pd

final_df['FlightDate'] = pd.to_datetime(final_df['DATE']).dt.date
origin_df = final_df.rename(columns={'Location': 'Origin', 'DailySnowDepth': 'DailySnowDepth_ORIGIN', 'DailySnowfall': 'DailySnowfall_ORIGIN'})
dest_df = final_df.rename(columns={'Location': 'Dest', 'DailySnowDepth': 'DailySnowDepth_DEST', 'DailySnowfall': 'DailySnowfall_DEST'})
flights_data['FlightDate'] = pd.to_datetime(flights_data['FlightDate']).dt.date

flights_data = flights_data.merge(origin_df[['FlightDate', 'Origin', 'DailySnowDepth_ORIGIN', 'DailySnowfall_ORIGIN']],
                                  on=['FlightDate', 'Origin'], how='left')

flights_data = flights_data.merge(dest_df[['FlightDate', 'Dest', 'DailySnowDepth_DEST', 'DailySnowfall_DEST']],
                                  on=['FlightDate', 'Dest'], how='left')

print(flights_data[['FlightDate', 'Origin', 'Dest', 'DailySnowDepth_ORIGIN', 'DailySnowfall_ORIGIN', 'DailySnowDepth_DEST', 'DailySnowfall_DEST']].head())
```

```python
print(final_df.columns)
```

```python
import matplotlib.pyplot as plt

final_df['DATE'] = pd.to_datetime(final_df['DATE'])
filtered_df = final_df[final_df['DATE'].dt.month.isin([1, 11, 12])]
filtered_df['DailySnowDepth'] = pd.to_numeric(filtered_df['DailySnowDepth'], errors='coerce')

filtered_df = filtered_df.dropna(subset=['DailySnowDepth'])

plt.figure(figsize=(10, 6))
plt.hist(filtered_df['DailySnowDepth'], bins=30, edgecolor='black')
plt.title('Distribution of Snow Depth in January, November, and December')
plt.xlabel('Daily Snow Depth')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

print(filtered_df['DailySnowDepth'].describe())
```

```python
flights_data[['DailySnowDepth_ORIGIN', 'DailySnowfall_ORIGIN', 'DailySnowDepth_DEST', 'DailySnowfall_DEST']] = flights_data[['DailySnowDepth_ORIGIN', 'DailySnowfall_ORIGIN', 'DailySnowDepth_DEST', 'DailySnowfall_DEST']].fillna(0)
```
